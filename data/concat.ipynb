```python
# example file name: delayedgerm_es:gs:hs:0.0:0.4:0.9:5:0.75.csv
# contact every csv file in the directory
# parse file name to get the parameters and add to the columns
#
# # Read the csv files in the directory
import os
import pandas as pd
```

```python
dist = False
csv_dir = '../data/GCN_ab/'
!pwd
```

```python

# Get the list of files in the directory
files = os.listdir(csv_dir)
files = [f for f in files if f.endswith('.csv') and 'delayedgerm' in f]

print(len(files))

# Initialize the dataframe
df = pd.DataFrame()

all_data = []

# Read each file and add the parameters to the dataframe
for file in files:
    # Extract the parameters from the file name, separated by underscore
    params = file.split('.csv')[0].split('delayedgerm_')[1].split(':')
    # Read the csv file
    data = pd.read_csv(csv_dir + file)
    # Add the parameters to the dataframe
    if dist:
        data['fits'] = params[0]
        data['germs'] = params[1]
        data['hards'] = params[2]
        data['pGermBH'] = params[3]
        data['pGermWT'] = params[4]
        data['bankLength'] = params[5]
        data['fmean'] = params[6]
        data['fvar'] = params[7]
        data['site'] = params[8]
    else:
        data['envs'] = params[0]
        data['germs'] = params[1]
        data['hards'] = params[2]
        data['pA'] = params[3]
        data['pGermBH'] = params[4]
        data['pGermWT'] = params[5]
        data['bankLength'] = params[6]
        data['wA'] = params[7]
        data['wB'] = params[8]
        data['site'] = params[9]
    # Append the data
    all_data.append(data)

df = pd.concat(all_data)

# # Write the dataframe to a new csv file
df.to_csv(csv_dir + 'data_maxn=1000.csv', index=False)
```
